```python
# Bar graph to check total cases in particular states 
plt.figure(figsize=(5,8))
sns.barplot(data=Df,x='Total Cases',y='State/UTs')
plt.show()
```

```python
df= pd.DataFrame(Df)
```

```python
# bar graph 
plt.figure(figsize=(20,10))
plt.bar(df['State/UTs'],df['Total Cases'])
plt.xlabel('State/UTs')
plt.ylabel('Total Cases')
plt.title("Covid")
plt.show()
```

```python
plt.scatter(df,x='Active Ratio(%)',y='Death Ratio(%)',color=df['State/UTs'])
plt.show()
```

```python
plt.scatter(df['Active Ratio (%)'], df['Death Ratio (%)'])
plt.xlabel('Active Ratio')
plt.ylabel('Death Ratio')
plt.title('Active Ratio vs. Death Ratio')
plt.show()
```
